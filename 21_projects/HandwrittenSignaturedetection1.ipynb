```python
# âœ… YOLOv8 Handwritten Signature Detection - Colab Ready Notebook

# ----------------------
# ğŸ› ï¸ Step 1: Install Dependencies
# ----------------------
!pip install ultralytics --quiet

# ----------------------
# ğŸ“ Step 2: Upload and Unzip Dataset
# ----------------------
from google.colab import files
import zipfile
import os

uploaded = files.upload()  # Upload the Roboflow-exported ZIP (YOLO format)

zip_filename = list(uploaded.keys())[0]  # Get uploaded file name
dataset_folder = "signature_dataset_1"

with zipfile.ZipFile(zip_filename, 'r') as zip_ref:
    zip_ref.extractall(dataset_folder)

print(f"âœ… Dataset extracted to: {dataset_folder}")

# ----------------------
# âœ… Step 3: Train YOLOv8s Model
# ----------------------
from ultralytics import YOLO

model = YOLO("yolov8s.pt")

model.train(
    data=f"{dataset_folder}/data.yaml",  # Update path to your dataset yaml
    epochs=50,
    imgsz=416,
    batch=8
)

# ----------------------
# ğŸ” Step 4: Run Inference on Test Image
# ----------------------
from google.colab import files
uploaded = files.upload()  # Upload a test image (e.g., scanned document)

import os
img_path = list(uploaded.keys())[0]

# Predict and save results
results = model.predict(source=img_path, save=True, conf=0.5)

# ----------------------
# ğŸ“Š Step 5: Print Detection Info
# ----------------------
for r in results:
    for box in r.boxes:
        cls_id = int(box.cls)
        conf = float(box.conf)
        coords = [round(x, 2) for x in box.xyxy[0].tolist()]
        print(f"\nâœ… Detected: {model.names[cls_id]}\nConfidence: {conf:.2f}\nBox: {coords}")

# ----------------------
# ğŸ–¼ï¸ Show Output Image
# ----------------------
from IPython.display import Image, display

output_dir = r.save_dir if hasattr(r, 'save_dir') else 'runs/detect/predict'
output_image = os.path.join(output_dir, img_path)
display(Image(filename=output_image))
```
